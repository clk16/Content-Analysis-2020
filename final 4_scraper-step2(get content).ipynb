```python
def GetOneRow(url,name0):
    header = set(ua.random)
    headers={"User-Agent" :ua.random,
    "Accept" : "text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8",
    "Accept-Language" : "en-us",
    "Connection" : "keep-alive",
    "Accept-Charset" : "GB2312,utf-8;q=0.7,*;q=0.7"}
    personinfo = requests.get(url, timeout=10,headers=headers)
    personsoup= bs4.BeautifulSoup(personinfo.text, 'html.parser')
    Age=re.findall("<tr>\s*<th>\s*Age\s*</th>\s*<td>\s*(\d*)\s*</td>\s*</tr>",str(personsoup))
    Occupation=re.findall("<tr>\s*<th>\s*Occupation\s*</th>\s*<td>\s*(.*)\s*</td>\s*</tr>",str(personsoup))
    Gender=re.findall("<tr>\s*<th>\s*Gender\s*</th>\s*<td>\s*(.*)\s*</td>\s*</tr>",str(personsoup))
    Ethnicity=re.findall("<tr>\s*<th>\s*Ethnicity\s*</th>\s*<td>\s*(.*)\s*</td>\s*</tr>",str(personsoup))
    Born=re.findall("<tr>\s*<th>\s*Born\s*</th>\s*<td>\s*(.*)\s*</td>\s*</tr>",str(personsoup))
    Residence=re.findall("<tr>\s*<th>\s*Residence\s*</th>\s*<td>\s*(.*)<*\s*</td>\s*</tr>",str(personsoup))
    CauseOfDeath=re.findall("<tr>\s*<th>\s*Cause of death\s*</th>\s*<td>\s*(.*)\s*</td>\s*</tr>",str(personsoup))
    nickname=re.findall("<tr>\s*<th>\s*Nickname\(s\)\s*</th>\s*<td>\s*(.*)\s*</td>\s*</tr>",str(personsoup))
    image=re.findall('<img.*data-src=?"?(https://.*?)?"',str(personsoup))[1]
    Name=name0
    actor=re.findall('portrayed by'+'(.*\s.*)\s.*',str(personsoup))
    newrow={'name':Name,'actor':actor,'nickname':nickname,'ethnicity':Ethnicity,'gender':Gender,'born':Born,'age':Age,'occupation':Occupation,'residence':Residence,'cause of death':CauseOfDeath,'image':image}
    return newrow
```

```python
os.chdir("D:/数据/美国data/soap opera")
df=pd.read_excel("characters.xlsx",Sheet='sheet1')
import nltk
```

```python
target=[]
for i in range(0,len(df)):
    newact=' '.join(nltk.word_tokenize(df['actor'][i].strip('[').strip(']').strip('\'').strip(" ").strip('actress').strip('actor')[0:20])[0:2])
    target.append(newact)
df['actor']=target
```

```python
nickname=[]
for i in range(0,len(df)):
    newnick=re.sub('<.*>','',df['nickname'][i]).strip('[').strip(']').strip("'").strip(' ')
    nickname.append(newnick)
df['nickname']=nickname
```

```python
eth=[]
for i in range(0,len(df)):
    neweth=re.sub('<.*>','',df['ethnicity'][i]).strip('[').strip(']').strip("'").strip(' ')
    eth.append(neweth)
df['ethnicity']=eth
df
```

```python
gender=[]
for i in range(0,len(df)):
    newgender=re.sub('<.*>','',df['gender'][i]).strip('[').strip(']').strip("'").strip(' ')
    gender.append(newgender)
df['gender']=gender
df
```

```python
born=[]
for i in range(0,len(df)):
    newborn=re.sub('<.*>','',df['born'][i]).strip('[').strip(']').strip("'").strip(' ')
    born.append(newborn)
df['born']=born
df
```

```python
age=[]
for i in range(0,len(df)):
    newage=re.sub('<.*>','',df['age'][i]).strip('[').strip(']').strip("'").strip(' ')
    age.append(newage)
df['age']=age
df
```

```python
occu=[]
for i in range(0,len(df)):
    newoccu=re.sub('<.*?>','',df['occupation'][i]).strip('[').strip(']').strip("'").strip(' ')
    occu.append(newoccu)
df['occupation']=occu
df
```

```python
resi=[]
for i in range(0,len(df)):
    newresi=re.sub('<.*?>','',df['residence'][i]).strip('[').strip(']').strip("'").strip(' ')
    resi.append(newresi)
df['residence']=resi
df
```

```python
cause=[]
for i in range(0,len(df)):
    newcause=re.sub('<.*?>','',df['cause of death'][i]).strip('[').strip(']').strip("'").strip(' ')
    cause.append(newcause)
df['cause of death']=cause
df
```

```python
image=[]
for i in range(0,len(df)):
    try:
        newimage=re.sub('<.*?>','',df['image'][i]).strip('[').strip(']').strip("'").strip(' ')
        image.append(newimage)
    except:
        image.append(df['image'][i])
df['image']=image
df
```
