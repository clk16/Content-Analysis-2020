```python
import os
os.chdir("D:/SOAP")
import pandas as pd
import numpy as np
import zipfile
import spacy
nlp = spacy.load("en")
import re
```

```python
corpus_name = "D:/SOAP"
def loadcorpus(corpus_name, corpus_style="text"):
    texts_raw = {}
    for file in os.listdir(corpus_name + "/"):
        if corpus_style in file:
            print(file)
            zfile = zipfile.ZipFile(corpus_name + "/" + file)
            for file in zfile.namelist():
                texts_raw[file] = []
                with zfile.open(file) as f:
                    for line in f:
                        texts_raw[file].append(line)
    return texts_raw
raw=loadcorpus(corpus_name)
```

```python
def clean_raw_text(raw_texts):
    clean_texts = []
    for text in raw_texts:
        try:
            text = text.decode("utf-8")
            clean_text = text.replace(" \'m", "'m").replace(" \'ll", "'ll").replace(" \'re", "'re").replace(" \'s", "'s").replace(" \'re", "'re").replace(" n\'t", "n't").replace(" \'ve", "'ve").replace(" /'d", "'d").replace(" \'ll", "'ll").replace(" \'d", "'d").replace(" \'m", "'m").replace(" \'s", "'s").replace(" \'t", "'t")
            clean_texts.append(clean_text)
        except AttributeError:
            print("ERROR CLEANING")
            print(text)
            continue
        except UnicodeDecodeError:
            print("Unicode Error, Skip")
            continue
    return clean_texts
```

```python
contentDF=pd.DataFrame(columns=['filmid','content'])
for i in raw:
    cleaned=clean_raw_text(raw[i])
    for j in range(0,len(cleaned)):
        try:
            filmid=re.findall("@@(\d+)",str(raw[i][j]))[0]
            content=re.findall("@@\d+(.*)",str(raw[i][j]))[0]
            contentDF=contentDF.append({'filmid':filmid,'content':content},ignore_index=True)
        except:
            print(str(raw[i][j][0:5])+' is wrong')
```

```python
contentDF['filmid']=contentDF['filmid'].astype(int)
```

```python
import os
os.chdir("D:/SOAP/")
basic=pd.read_excel("basic.xlsx",Sheet='sheet1')
basic
```

```python
result=pd.merge(basic,contentDF,how='left',left_on='textID',right_on='filmid')
```
