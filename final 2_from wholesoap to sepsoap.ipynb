```python
shownamedict=dict()
for i in set(list(namedf['show'])):
    targetdf=namedf[namedf['show']==i]
    namelist=list(targetdf['name'])
    shownamedict[i]=namelist
shownamedict
```

```python
def extract(sentence):
    if len(re.findall('@!(.*?):*\s',sentence))==2:
        speaker=re.findall('@!(.*?):*\s',sentence)[0]
        other=re.findall('@!(.*?):*\s',sentence)[1]
    else:
        speaker='NA'
        other='NA'
    if len(re.findall('@!.*?\s(.*?)@!.*?\s',sentence))>0:
        word=re.findall('@!.*?\s(.*?)@!.*?\s',sentence)[0].strip()
    else:
        word='NA'
    return [speaker,other,word] 
def gettext(i):
    df0=df.iloc[i]
    textid=df0['textID']
    text=df0['content']
    namelist=shownamedict[df0['show']]
    progress=i/len(df)*100
    print('progress: '+str(progress))
    try:
        sentences=regex.findall('@!.*?\s?.*?@!.*?\s',text,overlapped=True)
    except:
        sentences=[]
    return [extract(sentence) for sentence in sentences]
items=[x for x in range(0,len(df))]
import time
from multiprocessing.dummy import Pool as ThreadPool
if (__name__ == '__main__'):
    pool = ThreadPool(processes=12)
    results=pool.map(gettext, items)
    pool.close()
    pool.join()
```

```python
df['wordflow']=results
```

```python
def prowf(showname,textid,wordflow):
    speaker=[x[0] for x in wordflow if x is not None]
    listener=[x[1] for x in wordflow if x is not None]
    words=[x[2] for x in wordflow if x is not None]
    show=[showname]*len(speaker)
    textid=[textid]*len(speaker)
    newdf=pd.DataFrame({'show':show,'speaker':speaker,'listener':listener,'textID':textid,'words':words})
    return newdf
    
for i in range(0,len(df)):
    progress=i/len(df)*100
    print(progress)
    df0=df.iloc[i]
    showname=df0['show']
    wordflow=df0['wordflow']
    textid=df0['textID']
    newdf=prowf(showname,textid,wordflow)
    sodf=sodf.append(newdf)

sodf
```
